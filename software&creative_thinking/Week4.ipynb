if문 예제

```python
age = int(input('나이를 입력하세요 : '))
if age < 20:
  print('주류 구매가 불가능합니다')
else:
  print('주류 구매가 가능합니다')
```

```python
date = int(input('오늘 날짜를 입력하세요 : '))
if date % 2 == 0:
  print('짝수 차량만 운행 가능합니다')
else:
  print('홀수 차량만 운행 가능합니다')
```

```python
age = int(input('나이를 입력하세요 : '))
if age >= 20:
    print('성인입니다. 입장료는 5000원 입니다')
    price = 5000
else:
  print('성인이 아닙니다. 청소년 및 아동 요금은 2000원 입니다')
  price = 2000

cash = int(input("입장료를 투입하세요 : "))

if cash > price:
  print((cash - price), '원을 거슬러 드립니다.')
elif cash < price:
  print((price - cash), '원이 부족합니다.')
else:
  print('입장권이 출력됩니다')
```

```python
score = int(input('성적을 입력하세요 : '))

if score >= 90:
  print('A학점입니다')
elif score >= 80:
  print('B학점입니다')
elif score >= 70:
  print('C학점입니다')
elif score >= 60:
  print('D학점입니다')
else:
  print('F학점입니다')
```

```python
score = int(input('성적을 입력하세요 : '))

if score >= 90:
  str = 'A'
elif score >= 80:
  str = 'B'
elif score >= 70:
  str = 'C'
elif score >= 60:
  str = 'D'
else:
  str = 'F'

print(str, '학점입니다')
```

중첩 if

```python
date = int(input('오늘 날짜를 입력하세요 : '))
if date <= 31:
  if date % 2 == 0:
    print('짝수 차량만 운행 가능합니다')
  else:
    print('홀수 차량만 운행 가능합니다')
else:
  print('잘못 입력하셨습니다')
```

```python
score = int(input('성적을 입력하세요 : '))

if score >= 90:
  str = 'A'
else:
  if score >= 80:
    str = 'B'
  else:
    if score >= 70:
      str = 'C'
    else:
      if score >= 60:
        str = 'D'
      else:
        str = 'F'
print(str, '학점입니다')
```

```python
score = int(input('성적을 입력하세요 : '))

if score >= 60:
  if score >= 70:
    if score >= 80:
      if score > 90:
        str = 'A'
      else:
        str = 'B'
    else:
      str = 'C'
  else:
    str = 'D'
else:
  str = 'F'
print(str, '학점입니다')
```

```python
sum = 0
for i in range(1, 101):
  if i % 2 == 1:
    continue
  else:
    sum += i
  print(i, end = '-')

print('\n1 ~ 100 사이의 짝수들의 합은', sum)
```

```python
import random

roll = 0

while True:
  dice1 = random.randint(1, 6)
  dice2 = random.randint(1, 6)
  dice3 = random.randint(1, 6)

  roll += 1

  if (dice1 == dice2 and dice2 == dice3):
    print('주사위 숫자', dice1, '(으)로 모두 일치합니다. 던진 횟수 :', roll)
    break
```

```python
#15 숫자 맞추기 게임

import random

num = random.randint(1, 50)
guessTime = 0   #시도횟수

while True:
  i = input('숫자를 추측해보세요 : ')
  guessNum = int(i)
  guessTime += 1

  if (num > guessNum):
    print('추측한 수가 작습니다')
  elif (num < guessNum):
    print('추측한 수가 큽니다')
  else:
    print('정답입니다. 시도횟수 : ', guessTime)
    break
```

```python
spending = []
sum = 0

for i in range(6):
  print('{}월 지출 : ' .format(i + 1), end='')
  a = int(input())
  spending.append(a)
  sum += spending[i]

print(spending)
print('상반기 지출 총 합 : ', sum)
print(type(spending))
print(len(spending))
```

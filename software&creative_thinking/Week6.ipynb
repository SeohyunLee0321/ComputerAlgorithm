```python
#이차원 리스트 다루기 예제 1
spending = [['20210310', '커피', 3500],
            ['20210327', '영화', 15000],
            ['20210402', '커피', 5000],
            ['20210417', '책', 18000],
            ['20210427', '선물', 50000],
            ['20210505', '선물', 30000],
            ['20210508', '선물', 100000],
            ['20210520', '옷', 60000],
            ['20210528', '커피', 4000]]

#커피 항목 총 지출 금액 계산
coffee = 0
march = 0

for row in spending:
  if (row[1] == '커피'):
    coffee += row[2]  #리스트 마지막이기 때문에 row[-1]과 row[2]와 같음

#3월 총 지출 금액 계산
  date = row[0]
  if (date[4:6] == '03'):
    march += row[2]

print('커피 금액', coffee)
print('3월 금액', march)
```

```python
#이차원 리스트 다루기 예제 2
import random

bookList=[['역사탐험대'],
          ['파이썬'],
          ['학습법'],
          ['영단어'],
          ['여행에세이'],
          ['삼국지']]

max = 0
index = 0

for row in bookList:
  row.append(random.randint(15000, 25000))
  print(row)
  
  if (max < row[-1]):
    max = row[-1]

for i in range(len(bookList)):
  if (bookList[i][1] == max):
    index = i
    break

print('가장 비싼 책 :', bookList[index][0])
```

```python
midterm = {'Tom':43,
           'Mike':82,
           'Jamie':76,
           'John':61,
           'Mary':94}

for name in midterm.keys():
  print(name, end='')

print()

for score in midterm.values():
  print(score, end='')

print()

for name in midterm.keys():
  print(name, midterm[name], end='')

print()

for name, score in midterm.items():
  print(name, score, end='')

print()

for name, score in midterm.items():
  if score >= 70:
    print(name)
```

```python
import operator
store = {'삼각김밥':22, '육개장큰사발':30, '핫바':15, '구운계란':10, '우유':25, '생수':15}

for name, num in store.items():
  if num <= 20:
    print(name, end=' ')
    store[name] += 20
print()
print(store)


#print(sorted(menu.items(),key=operator.itemgetter(1),reverse=True))

store = dict(sorted(store.items(), key=operator.itemgetter(1), reverse=True))

for name in store.keys():
  print(name, end=' ')
```

```python
startInt = int(input('시작값:'))
endInt = int(input("끝"))
incValue = int(input('증가'))
sum = startInt

while True:
  print(startInt, end='')
  sum += startInt
  if startInt > endInt-incValue:
    print('=', end='')
    break
  else:
    print('+', end='')
    startInt += incValue
print(sum)
```

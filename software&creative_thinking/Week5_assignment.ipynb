```python
#14 [3] 구구단을 외우자! 게임

import random

print('\n#14 [3] 구구단을 외우자! 게임')
print('구구단을 외자!')

for i in range(5):
  num1 = random.randint(1, 9)
  num2 = random.randint(1, 9)
  
  num3 = int(input('{} * {} = ' .format(num1, num2)))

  if (num3 == num1 * num2):
    continue
  else:
    print('땡!')

print('끝')
```

```python
#15 숫자 맞추기 게임

import random

print('\n#15 숫자 맞추기 게임')

num = random.randint(1, 50)
guessTime = 0   #시도횟수

while True:
  i = input('숫자를 추측해보세요 : ')
  guessNum = int(i)
  guessTime += 1

  if (num > guessNum):
    print('추측한 수가 작습니다')
  elif (num < guessNum):
    print('추측한 수가 큽니다')
  else:
    print('정답입니다. 시도횟수 : ', guessTime)
    break
```

```python
#23~25 for문으로 리스트 다루기 예제2

import random

print('\n#23~25 for문으로 리스트 다루기 예제2')

food = ['피자', '치킨', '스테이크', '된장찌개', '바나나', '파스타', '감자튀김']
print(food)
recommend = random.choice(food)
print('점심 메뉴를 추천해 드립니다.\n추천 음식은 바로바로 --> ', recommend)

food.sort()
print('\n', food, sep = '')
add = input('추가하고 싶은 음식이 있습니까? y/n : ')

while(True):
  if (add == 'y'):
    addFood = input('음식이름(그만 추가하려면 q) : ')
    if (addFood == 'q'):
      print('\n')
      break;
    else:
      food.append(addFood)
  elif (add == 'n'):
    break;

print(food)
findFood = input('찾는 음식은? : ')
print((food.index(findFood) + 1), " 번째에 있습니다.")
```
